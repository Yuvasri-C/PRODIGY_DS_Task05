```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Road.csv')
```

```python
data.head(5)
```

```python
#Distribution of accidents by road conditions
plt.figure(figsize=(10, 6))
sns.countplot(x='Road_surface_conditions', data=data)
plt.title('Distribution of Accidents by Road Conditions')
plt.xlabel('Road Conditions')
plt.ylabel('Count of Accidents')
plt.show()
```

```python
#Distribution of accidents by time of day
plt.figure(figsize=(10, 6))
sns.countplot(x='Time', data=data)
plt.title('Distribution of Accidents by Time of Day')
plt.xlabel('Time of Day')
plt.ylabel('Count of Accidents')
plt.show()
```

```python
#Distribution of accidents by weather conditions
plt.figure(figsize=(10, 6))
sns.countplot(x='Weather_conditions', data=data)
plt.title('Distribution of Accidents by Weather Conditions')
plt.xlabel('Weather Conditions')
plt.ylabel('Count of Accidents')
plt.show()
```
